```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.drop(['EmployeeCount','EmployeeNumber','MonthlyRate','StockOptionLevel','TrainingTimesLastYear','PercentSalaryHike','EducationField'],axis = 1)
```

```python
# print(df['JobInvolvement'].unique())
print(df['BusinessTravel'].unique())
```

```python
df['BusinessTravel'] = df['BusinessTravel'].astype(str)
```

```python
# df['BusinessTravel'] = df['BusinessTravel'].replace('Non-Travel' , 'Non_Travel') 
df['BusinessTravel'] = df['BusinessTravel'].str.replace('Non-Travel', 'Non_Travel')
```

```python
df.isnull().sum()
```

```python
df['BusinessTravel'] = df['BusinessTravel'].str.strip()
```

```python
unique_values = df['BusinessTravel'].unique()
print(unique_values)
```

Converting Data into Int

```python
#DOUBT
cat_var = df.select_dtypes(include=['object']).columns.to_list()
# Label Encode the categorical variables in the extended train dataset
df_train_final = df.copy()
for variable in cat_var:
    df_train_final[variable] = df_train_final[variable].astype('category').cat.codes
df_train_final

```

Splitting of Data 

```python
X = df_train_final[['Age','MonthlyIncome','BusinessTravel',
       'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'Gender',
       'JobInvolvement', 'JobRole',
       'JobSatisfaction', 'MaritalStatus',  'WorkLifeBalance',
       'NumCompaniesWorked',  'PerformanceRating'
       ,'OverTime', 'TotalWorkingYears',
      'YearsAtCompany', 'YearsInCurrentRole',
       'YearsSinceLastPromotion']]
y = df_train_final['Attrition']
```

```python
X
```

```python
X.shape ,y.shape
```

```python
X_train , X_test , y_train, y_test = train_test_split(X ,y , train_size=.9 , random_state=1)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
output = model.predict(X_test)
```

```python
output
```

Accuracy of the model

```python
model.score(X_test,y_test)
```

Function Creation for user

```python
def ibm(Age,MonthlyIncome,BusinessTravel,
       DistanceFromHome, Education, EnvironmentSatisfaction, Gender,
       JobInvolvement,JobRole,
       JobSatisfaction, MaritalStatus,WorkLifeBalance,
       NumCompaniesWorked, PerformanceRating
       ,OverTime,TotalWorkingYears,
      YearsAtCompany, YearsInCurrentRole,
       YearsSinceLastPromotion):
    input_data = pd.DataFrame({
        'Age': [Age],
        'MonthlyIncome': [MonthlyIncome],
        'BusinessTravel': [BusinessTravel],
        'DistanceFromHome': [DistanceFromHome],
        'Education': [Education],
        'EnvironmentSatisfaction': [EnvironmentSatisfaction],
        'Gender': [Gender],
        'JobInvolvement': [JobInvolvement],
        'JobRole': [JobRole],
        'JobSatisfaction': [JobSatisfaction],
        'MaritalStatus': [MaritalStatus],
        'WorkLifeBalance': [WorkLifeBalance],
        'NumCompaniesWorked': [NumCompaniesWorked],
        'PerformanceRating': [PerformanceRating],
        'OverTime': [OverTime],
        'TotalWorkingYears': [TotalWorkingYears],
        'YearsAtCompany': [YearsAtCompany],
        'YearsInCurrentRole': [YearsInCurrentRole],
        'YearsSinceLastPromotion': [YearsSinceLastPromotion],
    })
    out = model.predict(input_data)
    if out ==1:
        print('The employee is bound to leave the organization')
    else:
        print('The employee will not leave the organization')
    
```
